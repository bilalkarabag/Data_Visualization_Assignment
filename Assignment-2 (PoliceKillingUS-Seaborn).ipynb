## 2: Plot 12 most dangerous cities

```python
mdc = kill.city.value_counts()[:12]
```

```python
mdc.values
```

```python
mdc.index
```

```python
mdc
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=mdc.index, y=mdc.values, data=kill)
plt.title("Most Dangerous Cities")
plt.xticks(rotation = 45);
```

```python
sns.countplot(x="city" , order=kill.city.value_counts()[:12].index ,data=kill)
plt.title("Most Dangerous Cities")
plt.xticks(rotation = 45);
```
